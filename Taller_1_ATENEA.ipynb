6. Usando Socrata, Cargue un dataset que se encuentre en el portal de datos públicos de Colombia

El dataset a cargar muestra la ejecución del Presupuesto General de la Nación hecha por las diferntes entidades que componen al Gobierno Nacional central. Consta de 748 filas y 11 columnas que muestran la fuente de financiación, ercurso presupuestal, entidades, obligaciones, etc.

```python
cliente = Socrata('www.datos.gov.co', None)
results = cliente.get("xjxk-qhsc")
dfp = pd.DataFrame.from_records (results)
dfp.head ()
```

7. Realice las tareas de limpieza y transformación necesarias

```python
dfp.info()
```

```python
# Eliminación de datos que no se usarán
dfp = dfp.drop(['codigo_del_sector', 'codigo_entidad'], axis=1)
dfp
```

```python
# Se renombran las variables
dfp = dfp.rename(columns={
    'fuente_de_financiaci_n': 'fuente',
    'situacion_de_fondos': 'situacion',
    'recurso_presupuestal': 'recurso',
    'nombre_del_sector': 'sector',
    'nombre_entidad': 'entidad',
    'apropiaci_n_vigente': 'apropiacion',
    'compromisos': 'compromisos',
    'obligaciones': 'obligaciones',
    'pagos': 'pagos'
})
dfp
```

```python
#Revisar si hay registros duplicados
print(dfp.duplicated())
```

```python
# Transformar los datos
dfp['fuente'] = dfp['fuente'].astype(str)
dfp['situacion'] = dfp['situacion'].astype(str)
dfp['recurso'] = dfp['recurso'].astype(str)
dfp['sector'] = dfp['sector'].astype(str)
dfp['entidad'] = dfp['entidad'].astype(str)
dfp['apropiacion'] = dfp['apropiacion'].astype(float)
dfp['compromisos'] = dfp['compromisos'].astype(float)
dfp['obligaciones'] = dfp['obligaciones'].astype(float)
dfp['pagos'] = dfp['pagos'].astype(float)

dfp.info()
```

```python
# Verificar datos nulos
datos_faltantes = dfp.isnull().sum()
print(datos_faltantes)
```

```python
# Para hacer más simple el manejo de cifras, se dividirán los valores
def simplificar_cifra(valor):
    return valor / 1e12  # Dividir por 1 billón

dfp['apropiacion'] = dfp['apropiacion'].apply(simplificar_cifra)
dfp['compromisos'] = dfp['compromisos'].apply(simplificar_cifra)
dfp['obligaciones'] = dfp['obligaciones'].apply(simplificar_cifra)
dfp['pagos'] = dfp['pagos'].apply(simplificar_cifra)
```

```python
# Se agrupa y se resumen los datos
resumen = dfp.groupby('fuente').agg({
    'apropiacion': 'sum',
    'compromisos': 'sum',
    'obligaciones': 'sum',
    'pagos': 'sum'
})
print(resumen)
```

8. Presentar el dataset, mostrando datos estadísticos y gráficas

```python
dfp.describe()
```

```python
# Agrupar y resumir los datos por la variable 'fuente' y calcular la suma de 'apropiacion'
resumen_fuente = dfp.groupby('fuente')['apropiacion'].sum()

total_apropiacion = resumen_fuente.sum()

porcentaje_apropiacion = (resumen_fuente / total_apropiacion) * 100

plt.figure()
bar_plot = porcentaje_apropiacion.plot(kind='bar', color = "yellow")
plt.xlabel('Fuente de Financiación')
plt.ylabel('Porcentaje de Apropiación')
plt.title('Porcentaje de Apropiación por Fuente de Financiación')
plt.xticks(rotation = 0)

for i, v in enumerate(porcentaje_apropiacion):
    bar_plot.text(i, v + 1, f'{v:.2f}%', ha='center', va='bottom')

plt.show()
```

Esta gráfica muestra que cerca del 95 % de la financiación de las entidades a nivel nacional proviene de transferencias y asignaciones presupuestales realizadas desde la Nación hacia el Gobierno Nacional Central.

```python
plt.figure()
plt.pie(resumen_fuente, labels=resumen_fuente.index, autopct='%.1f%%')
plt.title('Proporción de Apropiación por Fuente de Financiación')

plt.show()
```

En este gráfico se puede apreciar en mayor medida el presupuesto usado en entidades del país.

```python
resumen_fuente = dfp.groupby('fuente')['compromisos', 'obligaciones', 'pagos'].sum()

# Calcular los porcentajes de cada categoría
porcentajes = resumen_fuente.apply(lambda x: x / x.sum() * 100, axis=1)

plt.figure()
resumen_fuente.plot(kind='bar', stacked=True)
plt.xlabel('Fuente de Financiación')
plt.ylabel('Valor')
plt.title('Distribución de Compromisos, Obligaciones y Pagos por Fuente de Financiamiento')
plt.xticks(rotation = 0)

for i in range(len(resumen_fuente)):
    y_offset = 0
    for j in range(len(resumen_fuente.columns)):
        value = round(porcentajes.iloc[i, j], 2)
        plt.text(i, resumen_fuente.iloc[i, j] / 2 + y_offset, f'{value}%', ha='center', fontsize=8)
        y_offset += resumen_fuente.iloc[i, j]

plt.show()
```

Ahora bien, ese presupuesto, con origen de la Nación o de Recursos propios, se divide los compromisos, pagos y obligaciones que cada entidad ha asumido. Ambas fuentes de financiación se dividen casi que en igual medida para estos aspectos, es decir, se usa la tercera parte del presupuesto para los compromisos (obligaciones contraídas para realizar un gasto o una inversión en el futuro), obligaciones (compromisos ya ocntraídos por el GNC y que se deben cumplir) y pagos (desembolsos hechos por el GNC para cumplir las obligaciones y compromisos).

```python
# Agrupar y resumir los datos y calcular la suma de las obligaciones
resumen_entidad = dfp.groupby('entidad')['obligaciones'].sum()

top_entidades = resumen_entidad.sort_values(ascending=False).head(10)

plt.figure()
plt.pie(top_entidades, labels=top_entidades.index, autopct='%.1f%%')
plt.title('Top 10 Entidades con Mayor Suma de Obligaciones')

plt.show()
```

En esta gráfica podemos ver las entidades del país con mayores obligaciones. Colombia, al ser un país con déficit en algunos de sus saldos fiscales (es decir, gastamos más de lo que nos ingresa) lleva gran parte de su presupuesto al pago de la deuda pública. Cabe destacar que otros de los rubros importantes en los que se lleva una gran obligación es en educación y salud, esto se puede deber a la reciente pandemia que llevó a la compra masiva de vacunas y la financiación de estudiantes en el país con matrícula cero durante la pandemia. Algo importante a resaltar, si se hace la comparativa con años anteriores, es la disminución en el Departamento Administrativo para la Prosperidad Social con la eliminación de algunos subsidios que se dieron durante la pandemia.

```python
# Agrupar y resumir los datos y calcular la suma de las apropiaciones
resumen_entidad = dfp.groupby('entidad')['apropiacion'].sum()

top_entidades = resumen_entidad.sort_values(ascending=False).head(10)

plt.figure()
plt.pie(top_entidades, labels=top_entidades.index, autopct='%.1f%%')
plt.title('Top 10 Entidades con Mayor Suma de Apropiaciones')

plt.show()
```

Las apropiaciones hacen referencia a los recursos económicos asignados o destinados a diferentes actividades, programas o proyectos dentro de una entidad pública. De nuevo, los mayores recursos se destinan al pago de la deuda, salud y educación, que llegan a representar cerca del 62 % de las apropiaciones, lo que equivale a unos 214 billones de pesos.

```python
# Saber los tipos de recursos que más se usan para cada operación del PGN
recursos_counts = dfp['recurso'].value_counts()

plt.figure(figsize=(8, 6))
recursos_counts.plot(kind='bar', color = "purple")
plt.xlabel('Recursos')
plt.ylabel('Frecuencia')
plt.title('Distribución de Recursos')

# Mostrar el gráfico
plt.show()
```

En esta gráfica se observa que los recursos que se usan con más frecuencia para el funcionamiento de las entidades del país son Otros Recursos del Tesoro, Recursos Corrientes y los Recursos del Crédito Externo.

La situación de fondos se refiere al estado o condición de los recursos financieros disponibles en una entidad. En el DataSet hay dos tipos:

CSF: Corriente de Situación de Fondos. Esta categoría hace referencia a los recursos financieros disponibles en el presente, es decir, los fondos líquidos que se encuentran disponibles para ser utilizados en el corto plazo.

SSF: Situación de Fondos para Situaciones Especiales. Esta categoría se utiliza para identificar y gestionar los recursos financieros destinados a situaciones especiales o contingencias específicas.

Gran parte de financiación se usa en CSF para proyectos de inversión en educación, salud, infraestructura o programas sociales, como se observa en la tabla. Un punto a añadir es que, en caso de una emergencia de gran trascendencia nacional, no existen recursos habilitados, únicamente se pueden habilitar cuando se declare algún tipo de emergencia.
